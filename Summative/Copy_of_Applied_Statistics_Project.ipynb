## Task 3 - Probability Distributions üé≤

- **Identify variables that could follow specific probability distributions (e.g., Binomial, Normal). Calculate probabilities and expected values based on these distributions.**



```python
import scipy.stats
from scipy.stats import norm
```

# **Income**

```python
plt.hist(superstore.Income, bins=30, density=True, alpha=0.6, color='g')
plt.title('Probability Distribution Function of Income')
plt.xlabel('RMV Value')
plt.ylabel('Probability Density')

x = np.linspace(superstore['Income'].min(), superstore['Income'].max(), 100)
p = norm.pdf(x, np.mean(superstore.Income), np.std(superstore.Income))
plt.plot(x, p, 'k', linewidth=2)
expected_value = np.sum(x*p)
plt.legend(title = f'Expected Value {round(expected_value)}', loc=1)
plt.show()
```

# **NumDealsPurchases**

```python
plt.hist(superstore.NumDealsPurchases, bins=50, density=True, alpha=0.6, color='g')
plt.title('Probability Distribution Function of NumDealsPurchases')
plt.xlabel('RMV Value')
plt.ylabel('Probability Density')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(superstore.NumDealsPurchases), np.std(superstore.NumDealsPurchases))
plt.plot(x, p, 'k', linewidth=2)
expected_value = np.sum(x*p)
plt.legend(title = f'Expected Value {round(expected_value,2)}', loc=1)
plt.show()
```
